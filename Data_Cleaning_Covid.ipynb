**Intentando encontrar los datos de la localización**

```python
from geopy.geocoders import Nominatim

geolocator = Nominatim(user_agent="kaggle_learn")

import folium

```

```python
 len(China_W['Address_to_geo'])
```

```python
def Lat(row):
    try:
        
        return geolocator.geocode(row).point.latitude
    except:
        return 'No se pudo'

def Lon(row):
  try:
    return geolocator.geocode(row).point.longitude
  except:
    return 'No se pudo'

China_W['col']=range(1212)
China_W.set_index('col',inplace=True)

China_W['Latitude']=China_W['Address_to_geo']
China_W['Longitude']=China_W['Address_to_geo']

China_W['Latitude']=China_W['Latitude'].apply(Lat)
China_W['Longitude']=China_W['Longitude'].apply(Lon)





```

```python
China_W['Latitude'].loc[China_W['Latitude']=='No se pudo']=0
China_W['Longitude'].loc[China_W['Longitude']=='No se pudo']=0

Nuevo=gpd.GeoDataFrame(China_W, geometry=gpd.points_from_xy(China_W['Longitude'], China_W['Latitude']), crs=4326)

print(Nuevo.crs)
Nuevo
```

```python
from folium import Choropleth, Circle, Marker

m = folium.Map(location=[54, 15], tiles='openstreetmap', zoom_start=2)

# Add points to the map
for idx, row in Nuevo.iterrows():
    Marker([row['Latitude'], row['Longitude']], popup=str(row['Address_to_geo'])+ '\n' +str(row['Latitude'])+'--'+ str(row['Longitude'])).add_to(m)

# Display the map
m
```
